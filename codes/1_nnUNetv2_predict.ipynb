#### Coronary arteries

```python
import nibabel as nib
from totalsegmentator.python_api import totalsegmentator

!totalseg_set_license -l aca_AT069DK66G7SSW

def get_coronary_arteries_annotations(input_path, output_path):
    
    input_img = nib.load(input_path)
    output_img = totalsegmentator(input_img, task="coronary_arteries")
    nib.save(output_img, output_path)
    return print(f"Coronary arteries segmentation saved at path: {output_path}")

dir_path = f"D:\\Shruti\\IS_project\\nnunet_input_images"
ca_output_path = "D:\\Shruti\\IS_project\\preds_coronary_arteries"

for img_file in os.listdir(dir_path):
    
    try:
        
        get_coronary_arteries_annotations(input_path=os.path.join(dir_path, img_file),
                                          output_path=os.path.join(ca_output_path, img_file))
        
    except Exception as e:
        
        print(f"Failed to convert {img_file}: {str(e)}")
        
print("DONE .........")
```

#### Whole lung region

```python
def get_whole_lung_annotations(input_path, output_path):
    
    input_img = nib.load(input_path)
    output_img = totalsegmentator(input_img, roi_subset=['lung_upper_lobe_left',
                                                         'lung_upper_lobe_right',
                                                         'lung_lower_lobe_left',
                                                         'lung_lower_lobe_right',
                                                         'lung_middle_lobe_right'],
                                  ml=True)
    
    nib.save(output_img, output_path)
    return print(f"Whole lung segmentation saved at path: {output_path}")

dir_path = f"D:\\Shruti\\IS_project\\nnunet_input_images"
lung_output_path = "D:\\Shruti\\IS_project\\preds_whole_lung"

for img_file in os.listdir(dir_path):
    
    try:
        
        get_whole_lung_annotations(input_path=os.path.join(dir_path, img_file),
                                   output_path=os.path.join(lung_output_path, img_file))
        
    except Exception as e:
        
        print(f"Failed to convert {img_file}: {str(e)}")
        
print("DONE .........")
```
